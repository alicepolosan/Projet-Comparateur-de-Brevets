- liste des coûts année par année pour le BU + brevet européen --> entrées: brevet_liste (on prend la liste des pays du dernier dictionnaire)
- liste coûts cumulés pour BU + brevet européen

```python
def liste_annees (brevet_list):
    dic=brevet_list[-1]
    annee_depot=dic['annee_depot']
    annee_delivrance=dic['annee_delivrance']
    annee_fin= annee_depot + 20
    year_list= np.arange(annee_delivrance,annee_fin)
    return year_list

brevet_lt=[{'annee_depot': 2018, 'annee_delivrance': 2020, 'pays': ['Allemagne'] }]
print(liste_annees (brevet_lt), len(liste_annees (brevet_lt)))
```

```python
def couts_année (brevet_list):
    dic=brevet_list[-1]
    pays= dic['pays']
    data= df.loc[pays]
    indices=list(data.columns)[2:]
    #print(indices)
    data=data[indices]
    sommes= data.sum()
    couts_BE= sommes.to_numpy()
    couts_BU= np.sum(couts_BU_annees)
    return couts_BE, couts_BU

couts_année(brevet_lt)
```

```python
def couts_cumulés (brevet_list):
    dic=brevet_list[-1]
    pays= dic['pays']
    data= df.loc[pays]
    indices=list(data.columns)[2:]
    data=data[indices]
    sommes= data.sum()
    couts_BE=list(np.cumsum(sommes.to_numpy()))
    couts_BU=list(np.cumsum(np.array(couts_BU_annees)))
    return couts_BE, couts_BU

couts_cumulés(brevet_lt)
```
